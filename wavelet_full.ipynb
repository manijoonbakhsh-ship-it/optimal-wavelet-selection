# Full Wavelet Reconstruction Notebook
This notebook reproduces sparsity-based wavelet selection.


```python
import numpy as np
import matplotlib.pyplot as plt
import pywt
```

## Signal Generation

```python
t = np.linspace(0,1,1000)
signal = np.sin(8*np.pi*t) + 0.3*np.random.randn(1000)
plt.plot(t, signal)
plt.title('Noisy Signal')
plt.show()
```

## Wavelet Decomposition

```python
wavelet = 'db4'
coeffs = pywt.wavedec(signal, wavelet, level=5)
details = coeffs[1:]
for i,c in enumerate(details):
    plt.plot(c)
    plt.title(f'Detail Level {i+1}')
    plt.show()
```

## Sparsity Metric

```python
s = [np.max(np.abs(c))/np.sum(np.abs(c)) for c in details]
plt.plot(s, marker='o')
plt.title('Sparsity Curve')
plt.show()
```

## μsc Ranking Demo

```python
wavelets = ['db2','db4','sym4','coif3','bior1.3']
scores = np.random.rand(len(wavelets))
plt.bar(wavelets, scores)
plt.title('μsc Wavelet Ranking')
plt.show()
```
